```python
companylist = pd.read_csv('companylist.csv')
companydf = pd.DataFrame(companylist)
companydf
```

```python
tickers = companydf['Symbol'].tolist()
# tickers
```

```python
validtickers = {}
for i in range(len(tickers)):
    try:
        performanceData = requests.get("https://www.blackrock.com/tools/hackathon/performance", params= {'identifiers':tickers[i], 'includePositionReturns':'true'})
        performancejson = performanceData.json()
        sharpe = performancejson['resultMap']['RETURNS'][0]['latestPerf']['oneYearSharpeRatio']
        risk = performancejson['resultMap']['RETURNS'][0]['latestPerf']['oneYearRisk']
        validtickers[tickers[i]] = [sharpe, risk]
    except:
        continue
```
